```python
# --- Path setup ---
import sys, os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

# --- Imports ---
from my_ml_lib.nn import Value, get_all_nodes_and_edges
from visualize import draw_dot

```

```python
# Simple computation: (a*b) + (b + c)^2
a = Value(2.0, label='a')
b = Value(-3.0, label='b')
c = Value(10.0, label='c')

d = a * b
e = b + c
f = e ** 2
out = d + f

out.label = 'out'
print("Forward output:", out.data)
```

```python
out.backward()

for v in [a, b, c, d, e, f, out]:
    print(f"{v.label:5s} | data={v.data:7.4f} | grad={v.grad:7.4f}")
```
